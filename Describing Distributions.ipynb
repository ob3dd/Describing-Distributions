```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
plt.style.use(['seaborn-muted','fivethirtyeight'])
sns.set_context('talk')
```

```python
df = pd.read_csv("Medical Dataset/data (3).csv")
df.head()
```

## VitD_levels

```python
# plotting the KDE diagram
sns.histplot(data=df, x='VitD_levels', discrete=True, stat='density', kde = True);
```

### Mean

```python
# Find the mean
mean = df['VitD_levels'].mean()
mean
```

```python
# Adding a line to the mean
fig,ax = plt.subplots(figsize=(10,4))
sns.histplot(df['VitD_levels'], ax = ax, stat='probability')
ax.axvline(mean, color = 'red', linestyle='--');
ax.set_title('VitD_levels');
```

### Median

```python
# Find the median
median = df['VitD_levels'].median()
median
```

```python
# Adding a line to the median
col = 'VitD_levels'
feature = df[col]
mean = feature.mean()
median = feature.median()
fig,ax = plt.subplots(figsize=(10,4))
sns.histplot(feature ,ax=ax,stat='probability')
ax.axvline(mean, color = 'red',label=f'Mean = {mean:,.2f}')
ax.axvline(median, ls='--', color = 'green', label = f'Median = {median:,.2f}')
ax.set_title(f'{col}')
ax.legend();
```

### STD

```python
# Add a line for for +1 std from the mean
col = 'VitD_levels'
feature = df[col]
mean = feature.mean()
std = feature.std()
plus_one_std = mean + std
fig,ax = plt.subplots(figsize=(10,4))
sns.histplot(feature ,ax=ax,stat='probability')
ax.axvline(plus_one_std, color = 'black', linestyle='--', label=f'+1 std = {plus_one_std:,.2f}')
ax.set_title(f'{col}')
ax.legend();
```

```python
# Add a line for for -1 std from the mean
col = 'VitD_levels'
feature = df[col]
mean = feature.mean()
std = feature.std()
minus_one_std = mean - std
fig,ax = plt.subplots(figsize=(10,4))
sns.histplot(feature ,ax=ax,stat='probability')
ax.axvline(minus_one_std, color = 'black', linestyle='--', label = f'-1 std = {minus_one_std:,.2f}')
ax.set_title(f'{col}')
ax.legend();
```

### Highlight

```python
# Highlight the range between +1 and =1 std
col = 'VitD_levels'
feature = df[col]
mean = feature.mean()
median = feature.median()
std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std
fig,ax = plt.subplots(figsize=(10,4))
sns.histplot(feature ,ax=ax,stat='probability')
ax.axvline(plus_one_std, color = 'black',label=f'+1 std = {plus_one_std:,.2f}')
ax.axvline(minus_one_std, color = 'black', label = f'-1 std = {minus_one_std:,.2f}')
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow')
ax.set_title(f'{col}')
ax.legend();
```

```python
from scipy import stats
kurt = stats.kurtosis(df['VitD_levels'], fisher = False)
```
